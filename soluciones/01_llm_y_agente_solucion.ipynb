### Teoría breve
- **LLM (Large Language Model / Modelo de Lenguaje Grande):** modelo entrenado para comprender y generar texto.
- **Agente:** un LLM + instrucciones + **herramientas (tools)**. Sigue el ciclo *observar → decidir → actuar → responder*.

En este notebook: hablamos con el LLM, creamos una cadena con *prompt* y construimos nuestro **primer agente**.

```python
import sys, os
REPO = os.path.abspath("..")
if REPO not in sys.path:
    sys.path.insert(0, REPO)
from config import get_chat_model, get_embeddings
print("Entorno cargado ✔")
```

## Parte A · Hablar directo con el LLM
La forma más simple: mandar texto y recibir texto.

```python
llm = get_chat_model()
respuesta = llm.invoke("Explica en una sola frase qué es un agente de IA.")
print(respuesta.content)
```

## Parte B · Prompt template (plantilla de prompt)
Una plantilla separa las **instrucciones fijas** de la **entrada variable**.

```python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "Eres una experta en IA que explica con analogías simples."),
    ("user", "{pregunta}"),
])

cadena = prompt | llm

print(cadena.invoke({"pregunta": "¿Qué es un LLM?"}).content)
```

## Parte C · Tu primer agente
Un agente puede **decidir usar una herramienta**. Aquí definimos una tool simple.

```python
from langchain.agents import create_agent
from langchain_core.tools import tool

@tool
def longitud_texto(texto: str) -> int:
    """Devuelve cuántos caracteres tiene un texto."""
    return len(texto)

agente = create_agent(llm, tools=[longitud_texto],
                      system_prompt="Eres un asistente útil. Usa herramientas cuando ayuden.")

resultado = agente.invoke({"messages": [
    {"role": "user", "content": "¿Cuántos caracteres tiene la palabra 'inteligencia'? Usa la herramienta."}
]})
print(resultado["messages"][-1].content)
```
